```python
#1
import matplotlib.pylab as plt
def S(b,i,s,n):
    return -(b*i*s)/n
def I(b,i,s,n,y):
    return ((b*i*s)/n) - (y*i)
def R(y,i):
    return y*i
def thing(S,I,R,b,i,s,y,r,t):
    Ls = [s]
    Li = [i]
    Lr = [r]
    Lt = [0]
    n = s+i+r
    for a in range(1,t):
        Lt.append(a)
        Ls.append(Ls[a-1]+S(b,Li[a-1],Ls[a-1],n))
        Li.append(Li[a-1]+I(b,Li[a-1],Ls[a-1],n,y))
        Lr.append(Lr[a-1] + R(y,Li[a-1]))
    for bruh in range(len(Lr)):
        Greater = Lr[bruh]
        Lesser = Lr[bruh-1]
        if (Greater < Lesser):
            Greater = Lesser
    P = Greater / n
    plt.plot(Lt,Ls, label='Susceptible')
    plt.plot(Lt,Li, label='Infectious')
    plt.plot(Lt,Lr, label= P)
    plt.legend()
    print("Percentage of population that was ever sick was", sickly/n)
```

```python
thing(S,I,R,.5,1000,99000,.1,0,80)
```
